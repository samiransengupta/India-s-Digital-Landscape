```python
def generate_create_table(df, table_name, force_string_cols=None):
    """
    Generate SQL CREATE TABLE script for a Pandas DataFrame.
    
    Parameters:
    - df: pandas DataFrame
    - table_name: str, name of the SQL table
    - force_string_cols: list of column names (or cleaned names) to force as NVARCHAR
    
    Returns:
    - SQL string
    """
    if force_string_cols is None:
        force_string_cols = ['hhid', 'state_code']  # Add any ID-type columns here
    
    sql = f"CREATE TABLE {table_name} (\n"
    
    for col in df.columns:
        col_clean = col.strip().replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_").replace("-", "_")
        dtype = df[col].dtype
        
        # Force some columns as string
        if col_clean.lower() in [c.lower() for c in force_string_cols]:
            sql_type = "NVARCHAR(20)"
        elif dtype == object:
            sql_type = "NVARCHAR(MAX)"
        elif "int" in str(dtype):
            sql_type = "INT"
        elif "float" in str(dtype):
            sql_type = "FLOAT"
        elif "bool" in str(dtype):
            sql_type = "BIT"
        else:
            sql_type = "NVARCHAR(MAX)"
        
        sql += f"    [{col_clean}] {sql_type},\n"
    
    sql = sql.rstrip(",\n") + "\n);"
    return sql
```
